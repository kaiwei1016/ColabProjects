Python網路爬蟲: 本週最新電影資訊

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://movies.yahoo.com.tw/movie_thisweek.html'  #取得網頁的回應
response = requests.get(url)
data = BeautifulSoup(response.text , 'html.parser')  #解析html結構

movies = data.find_all("div", "release_movie_name")  #取得需要的html物件
movies_en = data.find_all("div", "en")
levels = data.find_all("div", "leveltext")
times = data.find_all("div", "release_movie_time")
result = []

for i, l in zip(movies, times):
    i = i.text.replace(" ","").split("\n")
    l = l.text.replace(" ","").split("\n")
    result.append({"中文片名":i[2], "英文片名":i[5], "期待度":i[11], "上映日期":l[2]})  #把資料儲存到result串列

df = pd.DataFrame(result)  #建立成DataFrame
df
```
